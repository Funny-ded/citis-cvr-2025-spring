```python
import numpy as np
import cv2

cv2.namedWindow("rgb")
cv2.createTrackbar("h", "rgb", 0, 255, lambda i : i)
cv2.createTrackbar("s", "rgb", 0, 255, lambda i : i)
cv2.createTrackbar("v", "rgb", 0, 255, lambda i : i)

img = np.zeros((1000, 1000, 3), np.uint8)

while(True):
    h = cv2.getTrackbarPos("h", "rgb")
    s = cv2.getTrackbarPos("s", "rgb")
    v = cv2.getTrackbarPos("v", "rgb")

    img[:, :, 0] = h
    img[:, :, 1] = s
    img[:, :, 2] = v

    rgb = cv2.cvtColor(img, cv2.COLOR_HSV2BGR)
    
    print(img[100, 100, :], rgb[100, 100, :])

    cv2.imshow("orig img", img)
    cv2.imshow("converted to rgb", rgb)

    key = cv2.waitKey(70)# & 0xFF

    if (key == ord('q')):
        break

cv2.destroyAllWindows()
cv2.waitKey(30)
```

```python
import numpy as np
import cv2

cam = cv2.VideoCapture(0)

avg = np.zeros((1080, 1920, 3), np.uint8)

curr = []
smoothed = []

while(True):
    success, frame = cam.read()

    if (not success):
        break

    #frame = cv2.resize(frame, (960, 540))

    #avg = 0.94 * avg + 0.06 * frame
    avg = cv2.addWeighted(avg, 0.94, frame, 0.06, 0)
    
    curr.append(frame[500, 500, 0])
    smoothed.append(avg[500, 500, 0])
    
    diff = cv2.absdiff(frame, avg)
    foreground_mask = cv2.inRange(diff, (20, 20, 20), (255, 255, 255))
    
    cv2.imshow("avg", avg)
    cv2.imshow("diff", diff)
    cv2.imshow("foreground_mask", foreground_mask)

    key = cv2.waitKey(70)

    if (key == ord('q')):
        break

cam.release()
cv2.destroyAllWindows()
cv2.waitKey(30)
```

```python
import matplotlib.pyplot as plt

plt.plot(curr)
plt.plot(smoothed)
```
