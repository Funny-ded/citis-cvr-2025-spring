```python
import cv2
import time

cam = cv2.VideoCapture(0)

shift = 30

bpm = 120

#i = 0

start_time = time.time()

while (True):
    success, frame = cam.read()

    if (not success):
        break

    i = int((time.time() - start_time) / 60 * bpm)

    if (i % 4 == 0):
        frame = 255 - frame
    
    if (i % 4 == 1):
        frame[:, :, 2] = 20
    
    if (i % 4 == 2):
        frame[:, :, 1] -= 120

    h, w, _ = frame.shape

    if (i % 4 == 3):
        frame[:, 0 : w - shift, 2] = frame[:, shift : , 2]

    #i += 1

    cv2.imshow("window", frame)
    key = cv2.waitKey(30)# & 0xFF

    if (key == ord('q')):
        break

cam.release()
cv2.destroyAllWindows()
cv2.waitKey(30)
```

```python
import numpy as np
import cv2
#import time

def nothing(i):
    pass

cv2.namedWindow("concat")
cv2.createTrackbar("low_blue", "concat", 150, 255, nothing)

cam = cv2.VideoCapture(0)

while (True):
    success, frame = cam.read()

    frame = cv2.resize(frame, (960, 540))

    if (not success):
        break

    low_blue = cv2.getTrackbarPos("low_blue", "concat")

    #print(low_blue)

    mask1 = cv2.inRange(frame, (low_blue, 0, 0), (255, 155, 155))

    ksz = 7
    kernel = np.ones((ksz, ksz))
    eroded = cv2.erode(mask1, kernel)

    concat = np.concatenate((mask1, eroded), axis=1)

    #mask2 = cv2.inRange(frame, (100, 100, 180), (230, 240, 255))

    output = cv2.connectedComponentsWithStats(eroded)

    num_components = output[0]
    stats = output[2]

    for i in range(num_components):
        top    = stats[i, cv2.CC_STAT_TOP]
        left   = stats[i, cv2.CC_STAT_LEFT]
        width  = stats[i, cv2.CC_STAT_WIDTH]
        height = stats[i, cv2.CC_STAT_HEIGHT]
        area = stats[i, cv2.CC_STAT_AREA]

        if (area > 6000 and area < 30000):
            print(left)
            cv2.rectangle(frame, (left, top), (left + width, top + height), (123, 234, 234), 5)

    cv2.imshow("concat", concat)
    #cv2.imshow("mask1", mask1)
    #cv2.imshow("eroded", eroded)
    #cv2.imshow("mask2", mask2)

    #unified = cv2.bitwise_or(mask1, mask2)

    #cv2.imshow("mask", unified)

    cv2.imshow("window", frame)
    key = cv2.waitKey(30)# & 0xFF

    if (key == ord('q')):
        break

cam.release()
cv2.destroyAllWindows()
cv2.waitKey(30)
```
